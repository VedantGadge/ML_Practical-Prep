```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('home_data.csv')
df
```

We check correlation of features wrt to our target variable(price) and drop features which have weak correlation(<0.3)

```python
df.select_dtypes(include='number').corr()['price']
```

```python
df = df.drop(columns=df.select_dtypes(include='number').corr()['price'].abs().loc[lambda x: x < 0.3].index)
df
```

```python
df = df.select_dtypes(include='number')
df.shape
```

Slicing the dataset n setting the first column as y_ytain as it is our target variable , and the other 9 columns as x_train

```python
x_train , y_train = df.iloc[:,1:10] , df.iloc[:,0]
y_train.name
```

```python
#Converting to numpy arrays for computation
X = x_train.values
Y = y_train.values

#Initialising parameters
n_samples , n_features = X.shape
weights = np.zeros(n_features)
bias = 0
learning_rate = 0.001
epochs = 10000
cost_history = []

X_mean = X.mean(axis=0)
X_std = X.std(axis=0)
X = (X - X_mean) / X_std
```

```python
#Gradient Descent
for epoch in range(epochs):
    y_pred = np.dot(X,weights) + bias
    error = y_pred - Y

    #Compute gradients
    dw = (1/n_samples) * np.dot(X.T,error)
    db = (1/n_samples) * np.sum(error)

    #Update parameters
    weights -=learning_rate * dw
    bias -= learning_rate * db

    #Compute and store the cost(MSE)
    cost = (1/(2*n_samples)) * np.sum(error**2)
    cost_history.append(cost)

    if epoch%1000 == 0:
        print(f'Epoch {epoch} Cost: {cost}')

```

```python
# Predictions
y_pred = np.dot(X, weights) + bias

# Evaluation
mse = np.mean((Y - y_pred) ** 2)
print(f"Final MSE: {mse:.2f}")
```

```python
# Plot cost history
import matplotlib.pyplot as plt
plt.plot(cost_history)
plt.xlabel('Epoch')
plt.ylabel('Cost (MSE)')
plt.title('Cost vs. Epochs')
plt.show()
```

```python
# Calculate R² score
ss_total = np.sum((Y - Y.mean()) ** 2)
ss_res = np.sum((Y - y_pred) ** 2)
r2_score = 1 - (ss_res / ss_total)
print(f"R² Score: {r2_score:.4f}")
```

```python
plt.scatter(Y, y_pred, alpha=0.5)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price')
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], 'r--')  
plt.show()
```
